## Which data do we need?
    1. Table organizator:
    organizator_id | organizator_name | organizator_name_clean | organizator_email
    
    2. Table events:
    event_id | event_name | event_address | event_start_date | event_end_date | event_start_time | event_end_time
    
    3. Table users:
    user_id | username | user_email | user_password | user_account_created | user_last_login | user_profile_picture
    
    4. Table pickups:
    pickup_cities | pickup_zip_code
    
    5. Table journey:
    event_address | pickup_zip_code | user_id | journey_id | journey_empty_spaces | journey_car | event_id | journey_text | journey_date | event_date | journey_start_time | event_start_time
    
    6. Table reservation:
    journey_id | reservation_requested_seats | journey_empty_spaces | reservation_text | user_id | journey_user_id

```python
# --------------- creds --------------- 
import uuid # random unique IDs
import pandas as pd # Dataframes
from faker import Faker # create random data
from random import randint # random numbers for creation of test data
import sys
DATA_PATH = sys.path[0].replace("notebooks", "base_data")
```

```python
# --------------- table organizators ---------------
table_organizator = pd.DataFrame({
    "organizator_id" : ['96c55241-eede-4102-ab0b-4d25cadff77e'],
    "organizator_name" : ["Eintracht Frankfurt GmbH"],
    "organizator_name_clean": ["Eintracht Frankfurt"],
    "organizator_email": ["info@ef.de"]
})
table_organizator.to_csv(DATA_PATH + "\\table_organizator.csv", index=False)
table_organizator
```

```python
# --------------- table events ---------------
table_events = pd.DataFrame({
    "event_id" : ["2ab60824-b539-4a1f-ae1a-f7d94d2d55bb"],
    "event_name" : ["Vfl Wolfsburg vs. Eintracht Frankfurt"],
    "organizator_name_clean" : ["Eintracht Frankfurt"],
    "event_address" : ["In den Allerwiesen 1, 38446 Wolfsburg"],
    "event_start_date" : ["2020-06-01"],
    "event_end_date" : ["2020-06-01"],
    "event_start_time" : ["15:30:00"],
    "event_end_time" : ["18:00:00"]
})
table_events.to_csv(DATA_PATH + "\\table_events.csv", index=False)
table_events
```

```python
# --------------- table users ---------------
table_user = []
faker = Faker(["de_DE"])

for i in range(1, 21):
    table_user.append({
        "user_id": uuid.uuid4(),
        "user_email" : faker.email(),
        "username" : faker.user_name(),
        "user_password" : faker.password(),
        "user_account_created" : faker.date(),
        "user_last_login" : faker.date(),
        "user_profile_picture" : "default.jpg",
        "event_id": "2ab60824-b539-4a1f-ae1a-f7d94d2d55bb"
        })
    
table_user = pd.DataFrame(table_user)
table_user = table_user[["user_id", "username", "user_email", "user_password", "user_account_created", "user_last_login", "user_profile_picture", "event_id"]]
table_user.to_csv(DATA_PATH + "\\table_user.csv", index=False)
table_user
```

```python
# --------------- table_pickups ---------------
table_pickups = pd.read_csv("D:/TechTalents/project-tt-app/server/base_data/German-Zip-Codes.csv", sep = ";")
table_pickups = table_pickups[["Ort", "Plz"]]
table_pickups = table_pickups.rename(columns={"Ort" : "pickup_city", "Plz": "pickup_zip_code"})
table_pickups = table_pickups.dropna()
table_pickups["pickup_zip_code"] = table_pickups["pickup_zip_code"].astype(int)
table_pickups.to_csv(DATA_PATH + "\\table_pickups.csv", index=False)
table_pickups
```

```python
# --------------- table_journey ---------------
table_journey = []
car_manufactures = ["Abarth","Alfa Romeo","Aston Martin","Audi","Bentley","BMW","Bugatti","Cadillac","Chevrolet","Chrysler",
                    "Citroën","Dacia","Daewoo","Daihatsu","Dodge","Donkervoort","DS","Ferrari","Fiat","Fisker","Ford",
                    "Honda","Hummer","Hyundai","Infiniti","Iveco","Jaguar","Jeep","Kia","KTM","Lada","Lamborghini",
                    "Lancia","Land Rover","Landwind","Lexus","Lotus","Maserati","Maybach","Mazda","McLaren","Mercedes-Benz",
                    "MG","Mini","Mitsubishi","Morgan","Nissan","Opel","Peugeot","Porsche","Renault","Rolls-Royce","Rover",
                    "Saab","Seat","Skoda","Smart","SsangYong","Subaru","Suzuki","Tesla","Toyota","Volkswagen","Volvo"]
for i in range(1, 11):
    x = randint(0, len(table_events)-1)
    y = randint(0, len(table_pickups)-1)
    z = randint(0, len(table_user)-1)
    a = randint(0, len(car_manufactures)-1)
    table_journey.append({
        "event_address" : table_events["event_address"][x],
        "event_id" : table_events["event_id"][x],
        "event_start_date" : table_events["event_start_date"][x],
        "event_start_time" : table_events["event_start_time"][x],
        "pickup_zip_code" : table_pickups["pickup_city"][y],
        "user_id": table_user["user_id"][z],
        "journey_id": uuid.uuid4(),
        "journey_empty_spaces": randint(1, 5),
        "journey_car": car_manufactures[a],
        "journey_text": faker.text(),
        "journey_date": table_events["event_start_date"][x],
        "journey_start_time": faker.time()[0:5],
        "journey_money": randint(0, 30)
    })

table_journey = pd.DataFrame(table_journey)
table_journey.to_csv(DATA_PATH + "\\table_journey.csv", index=False)
table_journey
```

```python
# --------------- table reservation ---------------
table_reservation = []
for i in range(1, 21):
    x = randint(0, len(table_events)-1)
    table_reservation.append({
        "reservation_id": uuid.uuid4(),
        "journey_id" : table_journey["journey_id"][x],
        "reservation_requested_seats" : randint(1, 5),
        "journey_empty_spaces" : table_journey["journey_empty_spaces"][x],
        "reservation_text" : faker.text(),
        "user_id": table_user["user_id"][randint(0, len(table_user)-1)],
        "journey_user_id": table_journey["user_id"][x],
        "reservation_status": "pending"
    })
table_reservation = pd.DataFrame(table_reservation)
table_reservation = table_reservation[["reservation_id", "journey_id", "reservation_requested_seats",
                                       "journey_empty_spaces", "reservation_text",
                                       "user_id", "journey_user_id", "reservation_status"]]
table_reservation.to_csv(DATA_PATH + "\\table_reservation.csv", index=False)
table_reservation
```
