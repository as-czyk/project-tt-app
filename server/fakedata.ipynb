## Which data do we need?
    1. Table events:
    event_id | event_name | event_address | event_start_date | event_end_date | event_start_time | event_end_time
    
    2. Table users:
    user_prename | user_name | user_email | user_alias | user_password | ticket_id | user_id | event_id
    
    3. Table pickups:
    pickup_cities | pickup_zip_code
    
    4. Table journey:
    event_address | pickup_zip_code | user_id | journey_id | journey_empty_spaces | journey_car | event_id | journey_text | journey_date | event_date | journey_start_time | event_start_time
    
    5. Table tickets:
    event_id | ticket_id | event_date

```python
# --------------- creds --------------- 
import uuid # random unique IDs
import pandas as pd # Dataframes
from faker import Faker # create random data
from random import randint # random numbers for creation of test data
```

```python
# --------------- table events ---------------
table_events = pd.DataFrame({
    "event_id" : ["2ab60824-b539-4a1f-ae1a-f7d94d2d55bb",
                 "7be626b0-c1cd-4a26-a111-60afb13ba940",
                 "eff7e9ae-9c58-42e1-8835-0d51c33dc480",
                 "e6930b99-e8ca-49f8-9583-89f54366dc14"],
    "event_name" : ["Time Warp",
                   "MAYDAY",
                   "Rock am Ring",
                   "World Club Dome"],
    "event_address" : ["Xaver-Fuhr-Straße 101, 68163 Mannheim", # this format is not good!
                      "Rheinlanddamm 200, 44139 Dortmund",
                      "Otto-Flimm-Straße, 53520 Nürburg",
                      "Mörfelder Landstraße 362, 60528 Frankfurt am Main"],
    "event_start_date" : ["2020-04-04",
                         "2020-04-30",
                         "2020-06-05",
                         "2020-06-05"],
    "event_end_date" : ["2020-04-05",
                       "2020-05-01",
                       "2020-06-07",
                       "2020-06-07"],
    "event_start_time" : ["19:00:00",
                         "17:00:00",
                         "12:00:00",
                         "20:00:00"],
    "event_end_time" : ["14:00:00",
                       "12:00:00",
                       "20:00:00",
                       "18:00:00"]
})
table_events.to_csv("table_events.csv")
table_events
```

```python
# --------------- table tickets ---------------
table_tickets = []
for i in range(1, 401):
    x = randint(0, len(table_events)-1)
    table_tickets.append({
        "event_id" : table_events["event_id"][x],
        "ticket_id" : uuid.uuid4(),
        "event_start_date" : table_events["event_start_date"][x],
        "event_end_date" : table_events["event_end_date"][x]
    })
table_tickets = pd.DataFrame(table_tickets)
table_tickets = table_tickets[["event_id", "ticket_id", "event_start_date", "event_end_date"]]
table_tickets.to_csv("table_tickets.csv")
table_tickets
```

```python
# --------------- table users ---------------
table_user = []
faker = Faker(["de_DE"])

for i in range(1, 101):
    x = randint(0, len(table_tickets)-1)
    table_user.append({
        "user_id": uuid.uuid4(),
        "user_prename" : faker.first_name(),
        "user_name" : faker.last_name(),
        "user_email" : faker.email(),
        "user_alias" : faker.user_name(),
        "user_password" : faker.password(),
        "ticket_id" : table_tickets["ticket_id"][x],
        "event_id" : table_tickets["event_id"][x],
        })

table_user = pd.DataFrame(table_user)
table_user = table_user[["user_id", "user_prename", "user_name", "user_email", "user_alias", "user_password", "ticket_id", 
                        "event_id"]]
table_user.to_csv("table_user.csv")
table_user
```

```python
# --------------- table_pickups ---------------
table_pickups = pd.read_csv("D:/TechTalents/project-tt-app/server/German-Zip-Codes.csv", sep = ";")
table_pickups = table_pickups[["Ort", "Plz"]]
table_pickups = table_pickups.rename(columns={"Ort" : "pickup_city", "Plz": "pickup_zip_code"})
table_pickups = table_pickups.dropna()
table_pickups["pickup_zip_code"] = table_pickups["pickup_zip_code"].astype(int)
table_pickups.to_csv("table_pickups.csv")
table_pickups
```
